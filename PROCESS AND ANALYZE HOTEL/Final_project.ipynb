```python
#drop the company's column
df.drop('company', axis=1, inplace=True)
```

```python
plt.hist(df['agent'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Distribution of Agent Column')
plt.xlabel('Agent')
plt.ylabel('Frequency')
plt.show()
# a bit right skewed with outliers!
```

```python
#use the median to impute the missing values in agent column
df['agent'].fillna(df['agent'].median(), inplace=True)

plt.hist(df['country'].dropna(), bins=30, color='red', edgecolor='black')
plt.title('Distribution of Country Column')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.show()
```

```python
print(df['country'].unique())
print(df['country'].mode())
#mode is PRT
```

```python
#explore monthly treands

# Convert 'arrival_date_month' to a categorical data type for proper ordering
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=months_order, ordered=True)

# Create a bar plot of bookings per month
plt.figure(figsize=(12, 6))
sns.countplot(x='arrival_date_month', data=df, hue='is_canceled', palette='viridis')
plt.title('Bookings per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
# Create box plots for categorical features
categorical_features = ['market_segment', 'distribution_channel', 'reserved_room_type', 'deposit_type', 'customer_type']
for feature in categorical_features:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=feature, data=df, hue='is_canceled', palette='viridis')
    plt.title(f'{feature} vs. Booking Cancellation')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```
